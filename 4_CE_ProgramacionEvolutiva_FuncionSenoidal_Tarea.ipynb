$$ f(x) = A sin (B*x + C*\pi) $$
Entonces, el objetivo de esta actividad es encontrar los parámetros A, B y C que hacen que la función senoidal se ajuste a los puntos.

## Programación Evolutiva ##

### Representación
Cada especie debe contener la solución y los parámetros de mutación, esto es:
$$ < A_i, B_i, C_i, \alpha_{iA},\alpha_{iB},\alpha_{iC}> $$

```python
def especies_iniciales(n):
    A = np.random.uniform(-100,100,(n,1))
    B = np.random.uniform(-100,100,(n,1))
    C = np.random.uniform(-100,100,(n,1))
    alpha = np.random.uniform(1,10,(n,3))
    return np.concatenate((A,B,C,alpha),axis=1)
especies = especies_iniciales(20)
print(especies[0])
```

```python
def graficar_solucion_especie(esp):
    A,B,C = esp[0],esp[1],esp[2]
    Yc = A*np.sin(B*X+C*math.pi)
    plt.figure()
    plt.plot(X,Y,'.')
    plt.plot(X,Yc)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
graficar_solucion_especie(especies[0])
```
